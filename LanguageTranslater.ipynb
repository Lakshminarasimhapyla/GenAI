```python
import torch
import gradio as gr
from transformers import pipeline
```

```python
text_translator = pipeline("translation",model="facebook/nllb-200-distilled-600M",torch_dtype=torch.bfloat16)
```

```python
import json
with open('language.json','r') as file:
  language_data=json.load(file)
```

```python
languages= {entry['Language']:entry['FLORES-200 code'] for entry in language_data}
```

```python
def get_FLORES_code_from_language(language):
  return languages.get(language,None)
```

```python
def translate_text(text,source_language,target_language):
  src_code = get_FLORES_code_from_language(source_language)
  tgt_code= get_FLORES_code_from_language(target_language)

  if not src_code or not tgt_code:
    return " One of the language not supported."

  translation = text_translator(text , src_lang =src_code,tgt_lang=tgt_code)
  return translation[0]["translation_text"]
```

```python
demo = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Input text to translate", lines=6),
        gr.Dropdown(choices=list(languages.keys()), label="Select Source Language"),
        gr.Dropdown(choices=list(languages.keys()), label="Select Target Language")
    ],
    outputs=[gr.Textbox(label="Translated text", lines=4)],
    title="Language Translator",
    description="Translates Text between Multiple Languages"
)

demo.launch()
```
